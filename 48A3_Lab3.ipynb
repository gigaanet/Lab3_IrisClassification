```python
!pip install scikit-learn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Iris (2).csv')
data['Species'] = data['Species'].replace({'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3})
X = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = data['Species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

# Convert to binary classification for ROC curve
y_test_binary = (y_test == 1).astype(int)  # Example: consider class 1 as positive
lr_probs = lr.predict_proba(X_test)[:, 1]
lr_fpr, lr_tpr, _ = roc_curve(y_test_binary, lr_probs)
lr_auc = auc(lr_fpr, lr_tpr)

# SVM
svm = SVC(probability=True)
svm.fit(X_train, y_train)

# Convert to binary classification for ROC curve
svm_probs = svm.predict_proba(X_test)[:, 1]
svm_fpr, svm_tpr, _ = roc_curve(y_test_binary, svm_probs)
svm_auc = auc(svm_fpr, svm_tpr)

# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

# Convert to binary classification for ROC curve
dt_probs = dt.predict_proba(X_test)[:, 1]
dt_fpr, dt_tpr, _ = roc_curve(y_test_binary, dt_probs)
dt_auc = auc(dt_fpr, dt_tpr)

# Plot ROC curves
plt.figure(figsize=(8, 6))
plt.plot(lr_fpr, lr_tpr, label=f'Logistic Regression (AUC = {lr_auc:.2f})')
plt.plot(svm_fpr, svm_tpr, label=f'SVM (AUC = {svm_auc:.2f})')
plt.plot(dt_fpr, dt_tpr, label=f'Decision Tree (AUC = {dt_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc='lower right')
plt.show()
```
